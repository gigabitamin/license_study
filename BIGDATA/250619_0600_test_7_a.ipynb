```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(2))
test.head(2)
```

```python
# import sklearn.preprocessing
# dir(sklearn.preprocessing)
from sklearn.preprocessing import MinMaxScaler
# import scipy.stats
# dir(scipy.stats)
from scipy.stats import zscore
# import sklearn.metrics
# dir(sklearn.metrics)
from sklearn.metrics import mean_squared_error # mse -> sqrt 루트 씌워서 rmse (x ** 0.5) # rmse 는 분류 문제 -> 랜덤포레스트 분류
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
train.shape
test.shape
```

```python
target = train['이용금액']
train = train.drop(['이용금액', 'ID'], axis=1)
test = test.drop('ID', axis=1)

target.shape
train.shape
test.shape
```

```python
# train_n = train.select_dtypes(include='number')
# 년과 월은 스케일링이 불필요, number 대신 int64 로 변경
# train_n = train.select_dtypes(include='int64')
# 성능평가 결과 말도 안되는 수치 -> 날짜도 스케일링, 연월 구분없이 진행 or 원본 연월 그대로 살려서 진행
# (mean_squared_error(y_val, pred))**0.5 # 169841133.74905077
train_n = train.select_dtypes(include='number')
train_n[:1]
test_n = test.select_dtypes(include='number')
test_n[:1]
```

```python
x_train = train.drop('target', axis=1)
x_test = test.drop('target', axis=1)
y_train = train['target']
y_test = test['target']
```

```python
import statsmodels.api as sm
x_train = sm.add_constant(x_train)
x_test = sm.add_constant(x_test)
```

```python
model = sm.GLM(y_train, x_train, family=sm.families.Binomial())
result = model.fit()
result.summary()
```

```python
pred_proba = result.predict(x_test)
pred_proba
```

```python
pred =  (pred_proba >= 0.5)
pred
pred.astype('int')
```
