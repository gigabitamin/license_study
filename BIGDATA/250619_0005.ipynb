```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_test.csv")

display(x_train.head())
display(y_train.head())
```

```python
# 범주형 변수인데 적당히 많은 unique값을 가진 컬럼은 날린다.
drop_col = ['enrollee_id','city','company_type','experience']
x_train_drop = x_train.drop(columns = drop_col)
x_test_drop = x_test.drop(columns = drop_col)
```

```python
# import sklearn
# print(sklearn.__all__)
# import sklearn.model_selection
# print(dir(sklearn.model_selection))

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
x_train_dummies = pd.get_dummies(x_train_drop)
y = y_train['target'].astype('int')

x_test_dummies = pd.get_dummies(x_test_drop)
# train과 컬럼 순서 동일하게 하기 (더미화 하면서 순서대로 정렬을 이미 하기 때문에 오류가 난다면 해당 컬럼이 누락된것)
x_test_dummies = x_test_dummies[x_train_dummies.columns]
# print(help(train_test_split))
```

```python
X_train, X_validation, Y_train, Y_validation = train_test_split(x_train_dummies, y, test_size=0.33, random_state=42)
X_train.shape, X_validation.shape, Y_train.shape, Y_validation.shape
```

```python
rf = RandomForestClassifier(random_state =25)
rf.fit(X_train,Y_train)
```

```python
# import sklearn.metrics
# print(dir(sklearn.metrics))

from sklearn.metrics import accuracy_score , f1_score, recall_score, roc_auc_score ,precision_score

# model_score
predict_train_label = rf.predict(X_train)
predict_train_proba = rf.predict_proba(X_train)[:,1]

predict_validation_label = rf.predict(X_validation)
predict_validation_prob = rf.predict_proba(X_validation)[:,1]
```

```python
# 공통 전처리 대상 데이터프레임 합치기 (동일한 처리를 위해)
train = x_train.copy()
test = x_test.copy()
all_data = pd.concat([train, test], axis=0, ignore_index=True)
all_data.head()
all_data.info()
```

```python
# 결측치 처리

# 1. gender
all_data['gender'] = all_data['gender'].fillna('Unknown')

# 2. enrolled_university
all_data['enrolled_university'] = all_data['enrolled_university'].fillna('no_enrollment')

# 3. education_level
all_data['education_level'] = all_data['education_level'].fillna('Unknown')

# 4. major_discipline
all_data['major_discipline'] = all_data['major_discipline'].fillna('No Major')
```

```python
# 5. experience (문자 -> 숫자 변환 후 처리)
def clean_experience(val):
    if val == '>20':
        return 21
    elif val == '<1':
        return 0
    else:
        try:
            return int(val)
        except:
            return None

all_data['experience'] = all_data['experience'].apply(clean_experience)
all_data['experience'] = all_data['experience'].fillna(all_data['experience'].median())

all_data['experience']
```

```python
# 6. company_size
all_data['company_size'] = all_data['company_size'].fillna('Unknown')

# 7. company_type
all_data['company_type'] = all_data['company_type'].fillna('Unknown')

all_data.head(3)
```

```python
# 8. last_new_job
def clean_new_job(val):
    if val == '>4':
        return 5
    elif val == 'never':
        return 0
    else:
        try:
            return int(val)
        except:
            return None

all_data['last_new_job'] = all_data['last_new_job'].apply(clean_new_job)
all_data['last_new_job'] = all_data['last_new_job'].fillna(all_data['last_new_job'].median())

all_data.head()
```

### 1-3) CO(GT),NMHC(GT) 두 컬럼을 min-max normalize를 진행하고 각 컬럼의 모표준편차의 차이의 절댓값을 소숫점 이하 3째 자리까지 구하여라  
  
 

```python
import pandas as pd
df_o = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p1_3.csv')
df = df_o.copy()
df.head(5)
```

```python
df.info()
```

```python
import numpy as np

def min_max_nor(x):
    nmn = (x - x.mean()) / (x.max() - x.min())
    return nmn


df['CO(GT)'] = df['CO(GT)'].agg(min_max_nor)
df['NMHC(GT)'] = df['NMHC(GT)'].agg(min_max_nor)
# df

co_std = df['CO(GT)'].std(ddof=0)
nmhc_std = df['NMHC(GT)'].std(ddof=0)

round(abs(co_std - nmhc_std), 3)
```

```python
import numpy as np

def min_max_nor(x):
    nmn = (x - x.mean()) / (x.max() - x.min())
    return nmn


df['CO(GT)'] = df['CO(GT)'].agg(min_max_nor)
df['NMHC(GT)'] = df['NMHC(GT)'].agg(min_max_nor)
# df

def pop_std (arr):
    mean = arr.mean()
    var = sum((x - mean)**2 for x in arr) / len(arr)
    return var ** 0.5

round(abs(pop_std(df['CO(GT)']) - pop_std(df['NMHC(GT)'])),3)
```

```python
import pandas as pd

# 데이터 로드
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p2_test2.csv')

# 타겟 분리
y = train['Y']
train.drop(columns=['Y'], inplace=True)

# ID 제거
train.drop(columns=['ID'], inplace=True)
test.drop(columns=['ID'], inplace=True)

# 날짜 처리 (파생 변수 생성)
for df in [train, test]:
    df['Date'] = pd.to_datetime(df['Date'])
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['DayofWeek'] = df['Date'].dt.dayofweek
    df.drop(columns=['Date'], inplace=True)

# 범주형 처리 (One-Hot Encoding)
categorical_cols = ['C1', 'C2', 'C3', 'C4', 'Weekday']
train = pd.get_dummies(train, columns=categorical_cols)
test = pd.get_dummies(test, columns=categorical_cols)

# (주의) test에만 있고 train에 없는 컬럼 또는 반대가 있을 수 있음 → align으로 컬럼 정렬
train, test = train.align(test, join='left', axis=1, fill_value=0)

# 이제 train, test는 모두 숫자형 컬럼만 포함된 상태
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# 학습용 분리
X_train, X_val, y_train, y_val = train_test_split(train, y, test_size=0.2, random_state=42)

# 모델 학습
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# 검증
pred = model.predict(X_val)
print("MAE:", mean_absolute_error(y_val, pred))
```

### 3-4) PIQ를 예측하는 다중선형회귀 모형을 만들고 Brain의 회귀 계수를 구하여라

```python
import pandas as pd
import statsmodels.api as sm

# 데이터 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e8_p3_2.csv')

# 독립변수와 종속변수 분리
X = df[['Brain', 'Height', 'Weight']]
y = df['PIQ']

# 상수항 추가
X = sm.add_constant(X)

# 다중선형회귀 모형 적합
model = sm.OLS(y, X)
result = model.fit()

# Brain 변수의 회귀계수 출력
print("Brain 회귀계수:", result.params['Brain'])
```
